```python
import multiprocessing
import pandas as pd

# Define the map function
def map_func(record):
    # Check if the person is female and older than 40
    if record['Gender'] == 'Female' and record['Age'] > 40:
        # Return the hospital name with count 1 (for each female > 40)
        return (record['Hospital'], 1)
    else:
        # Return None if conditions are not met
        return None

# Define the reduce function
def reduce_func(mapped_data):
    hospital_visit_count = {}

    # Count the visits to each hospital
    for hospital, count in mapped_data:
        if hospital in hospital_visit_count:
            hospital_visit_count[hospital] += count
        else:
            hospital_visit_count[hospital] = count

    # Filter out hospitals visited by only one person
    unique_hospitals = {hospital: count for hospital, count in hospital_visit_count.items() if count == 1}

    return unique_hospitals

# Main function to run the MapReduce job
if __name__ == "__main__":
    # Load the dataset (assuming it's a CSV file)
    df = pd.read_csv('/content/healthcare_dataset.csv')  # Replace with actual file path

    # Apply the Map phase (in parallel)
    with multiprocessing.Pool() as pool:
        # Get the mapped results (female > 40 visiting hospitals)
        mapped_data = pool.map(map_func, df.to_dict(orient='records'))

    # Filter out None values from the map output
    mapped_data = [data for data in mapped_data if data is not None]

    # Apply the Reduce phase to get the unique hospitals
    unique_hospitals = reduce_func(mapped_data)

    # Print the final output
    print(f"Number of unique hospitals visited by females over 40: {len(unique_hospitals)}")
    print(f"Unique hospitals: {list(unique_hospitals.keys())}")
```
